```python
# Import Dependacies
from bs4 import BeautifulSoup as soup
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#Set up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

#Visit URL
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
#Setup BeautifulSoup
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
container = html_soup.findAll('div', id='news')
container
```

```python
# Loop through and gather all titles and summary from webpage
for row in container:
    titles =html_soup.findAll('div', class_='content_title')
    summaries=html_soup.findAll('div', class_='article_teaser_body')
    
doc_info = {
    'title':titles,
    'summary':summaries
}
    
print(doc_info)
```

```python
#Create a list of dictionaries to store the title-and-preview pairs
articles = []
for i in range(len(titles)):
    articles.append({'title': titles[i].text, 'preview': summaries[i].text})
    print(articles)
```

```python
browser.quit()
```
